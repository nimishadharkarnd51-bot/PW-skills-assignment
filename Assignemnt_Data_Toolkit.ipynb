#2 How does broadcasting work in NumPy
Answer - Broadcasting is a feature in NumPy that allows operations on arrays of different shapes.
NumPy automatically expands the smaller array so both arrays become compatible for element-wise operations.
If two arrays have different shapes, NumPy tries to match them from the last dimension.
A dimension can be made compatible when:

Both dimensions are equal

One of them is 1 → NumPy repeats (broadcasts) it automatically.
Example
Example 1: Adding scalar to array
import numpy as np

a = np.array([1, 2, 3])
b = 5

print(a + b)

Output
[6 7 8]


NumPy broadcasts 5 → [5, 5, 5] automatically.

Example 2: Adding a row vector to a matrix
import numpy as np

a = np.array([[1, 2, 3],
              [4, 5, 6]])

b = np.array([10, 20, 30])

print(a + b)

Output
[[11 22 33]
 [14 25 36]]


NumPy broadcasts b to match shape (2, 3).

#6 What are the differences between NumPy arrays and Python lists
Answer - NumPy arrays and Python lists differ in several important ways. A NumPy array stores data in a continuous block of memory, making it much faster and more efficient for numerical operations, while a Python list stores elements in separate memory locations, which slows down computation. NumPy arrays require all elements to have the same data type, whereas Python lists can hold mixed data types. NumPy also supports vectorized operations, meaning mathematical calculations can be performed on entire arrays without using loops, but Python lists require explicit iteration for such tasks. Because of these features, NumPy arrays use less memory and offer advanced mathematical functions, making them ideal for scientific and data-heavy applications, whereas Python lists are more general-purpose and flexible but less suitable for large-scale numerical computing.

#17 Explain the difference between apply() and map() in Pandas
Answer - Difference Between apply() and map() in Pandas
1. Purpose
map()

Works only on Series.

Used for element-wise transformation.

Ideal for simple value mapping or replacements.

apply()

Works on both Series and DataFrames.

Used for applying functions along rows or columns.

Supports complex operations on groups of values.

2. Level of Operation
map()

Applies a function to each element of a Series.

Example:

s.map(lambda x: x + 1)

apply()

On a Series → works on each element (similar to map()).

On a DataFrame → works on rows or columns as a whole.

Example:

df.apply(sum, axis=0)  # sum of each column
df.apply(sum, axis=1)  # sum of each row

3. Input Types
map() accepts:

Dictionary (for mapping)

Function (element-wise)

Series (alignment-based mapping)

apply() accepts:

Any Python function

Lambda functions

Built-in functions

Custom row/column operations

4. When to Use What
✔ Use map() when:

You want simple value replacement

You are working on a single Series

The transformation is strictly element-wise

Example:

s.map({'M': 'Male', 'F': 'Female'})

✔ Use apply() when:

The operation requires logic across rows or columns

You need more complex calculations

You are working with a DataFrame

Example:

df.apply(lambda row: row['A'] + row['B'], axis=1)

#21 Why is NumPy’s array slicing faster than Python’s list slicing
Answer - NumPy’s array slicing is much faster than Python’s list slicing because of the way NumPy stores and processes data in memory. The speed advantage comes from contiguous memory storage, vectorization, and views instead of copies.

Here’s the clear explanation:

⭐ Why NumPy Array Slicing Is Faster Than Python List Slicing
1. NumPy uses contiguous memory blocks

NumPy arrays store data in one continuous block of memory (similar to C arrays).
This allows:

Faster access

Better CPU cache utilization

Efficient low-level operations

Python lists store elements scattered across memory, pointing to objects; this slows down slicing.

2. Slicing in NumPy creates a view, not a copy

When you slice a NumPy array:

arr[1:5]


NumPy does not copy data—it creates a view that references the same memory.

This makes slicing:

Fast

Memory-efficient

Python lists always create a new list when sliced, which takes time and memory:

lst[1:5]  # new list created → slow

3. NumPy is written in optimized C code

NumPy operations (including slicing) are executed using:

C loops

Low-level memory pointers

Highly optimized routines

This avoids the overhead of Python-level loops and object handling.

4. NumPy’s fixed data types allow faster computation

NumPy arrays store elements of the same data type.
This uniformity:

Minimizes overhead

Allows fast vectorized operations

Enables CPU optimizations

Python lists store objects of different types, slowing down processing.

```python
#8 Create a line plot with multiple lines using Matplotlib
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 3, 5, 7, 9]
y3 = [2, 3, 4, 3, 2]

# Plotting multiple lines
plt.plot(x, y1, label='Line 1')
plt.plot(x, y2, label='Line 2')
plt.plot(x, y3, label='Line 3')

# Adding labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Multiple Line Plot')

# Show legend
plt.legend()

# Display the plot
plt.show()
```

```python
#9 Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold
import pandas as pd

# Create a sample DataFrame
data = {
    'Name': ['A', 'B', 'C', 'D', 'E'],
    'Score': [45, 72, 88, 53, 30]
}

df = pd.DataFrame(data)

# Define threshold
threshold = 50

# Filter rows where Score > threshold
filtered_df = df[df['Score'] > threshold]

print("Original DataFrame:")
print(df)

print("\nFiltered DataFrame (Score > 50):")
print(filtered_df)
```

```python
#Create a histogram using Seaborn to visualize a distribution
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
data = [12, 15, 14, 17, 20, 21, 19, 18, 16, 22, 25, 23, 24, 20, 18, 17]

# Create histogram
sns.histplot(data, bins=8, kde=True)

# Add title
plt.title("Distribution Histogram")

# Show plot
plt.show()
```

```python
#11 Perform matrix multiplication using NumPy
import numpy as np

# Define two matrices
A = np.array([[1, 2, 3],
              [4, 5, 6]])

B = np.array([[7, 8],
              [9, 10],
              [11, 12]])

# Matrix multiplication
result = np.dot(A, B)      # Method 1
# result = A @ B           # Method 2 (same output)

print("Matrix A:")
print(A)

print("\nMatrix B:")
print(B)

print("\nResult of A x B:")
print(result)
```

```python
#13 Create a 3D scatter plot using Plotly
import plotly.express as px
import pandas as pd

# Sample data
data = {
    'x': [1, 2, 3, 4, 5],
    'y': [10, 15, 13, 17, 14],
    'z': [5, 7, 6, 8, 9],
    'category': ['A', 'A', 'B', 'B', 'C']
}

df = pd.DataFrame(data)

# Create 3D scatter plot
fig = px.scatter_3d(df, x='x', y='y', z='z', color='category',
                    title='3D Scatter Plot')

fig.show()
```
