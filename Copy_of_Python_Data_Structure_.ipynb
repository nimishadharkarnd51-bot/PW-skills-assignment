2.Explain the difference between mutable and immutable data types with examples.
Answer - **Mutable Data Types**:
Mutable data types are those whose internal state or content can be changed after they are initialized. When you modify a mutable object, you are changing the object itself in memory, and any variables referencing that object will reflect the change.
Examples: Lists, dictionaries, and sets in Python.
**Immutable Data Types**:
Immutable data types are those whose values cannot be altered once they are created. Any operation that appears to "modify" an immutable object actually results in the creation of a new object with the new value, leaving the original object unchanged.
Examples: Integers, floats, strings, and tuples in Python.

8. What is a hash table, and how does it relate to dictionaries in Python?
Answer - A hash table is a data structure that stores data in key–value pairs and uses a hashing function to decide where each key should be placed in memory. When a key is inserted, Python computes its hash value—a numerical representation of the key—and uses this value to determine the index (or “bucket”) where the corresponding value will be stored. This allows extremely fast access because Python can go directly to the bucket rather than searching through all items.

Dictionaries in Python are built on top of hash tables, which is why dictionary operations like lookup, insertion, and deletion are very fast—usually O(1) time. Each key in a dictionary must be hashable so that Python can generate a stable hash value for it. Because of this structure, dictionaries can efficiently map keys to values, handle collisions, and resize themselves automatically as data grows. In short, a dictionary is Python’s implementation of a hash table, providing quick and efficient key-based data retrieval.

14. How does the “in” keyword work differently for lists and dictionaries?
Answer - The “in” keyword behaves differently for lists and dictionaries because it checks different types of information in each structure.

For lists, the in keyword checks whether a specific value exists in the list. Python searches through the list element by element, so the check is based on the list’s contents and takes linear time.

For dictionaries, the in keyword checks whether a specific key exists in the dictionary, not the value. It does not scan the values unless you explicitly ask it to using .values(). The check is very fast because dictionaries use hash tables, allowing quick lookup of keys.

So, in summary: in a list, in looks for a value by scanning through items, while in a dictionary, in checks for a key using fast hashing.

17. Describe the time complexity of accessing elements in a dictionary
Answer . In Python, dictionaries are implemented as hash tables, so the time complexity for accessing elements is:

 Average Case: O(1)

Accessing an element using a key (e.g., value = my_dict[key]) takes constant time.

Python computes the hash of the key and directly finds the value in the underlying hash table.

 Worst Case: O(n)

In rare cases where many keys hash to the same location (hash collisions), access can degrade to linear time.

This is uncommon because Python optimizes hashing and collision handling internally.

Access time complexity of a dictionary in Python = O(1) average case, O(n) worst case

19. Why are dictionaries considered unordered, and how does that affect data retrieval?
Answer - Dictionaries are considered unordered because they store items using a hash table, where the position of each key–value pair depends on its hash value rather than insertion index. As a result, dictionaries do not support indexing or positional access. Data is retrieved by key lookup, not by position, and this lookup is performed in O(1) average time using the hash of the key.

20. Explain the difference between a list and a dictionary in terms of data retrieval.
Answer - A list and a dictionary differ mainly in how data is retrieved. In a list, elements are accessed using their index positions, meaning you must know the exact location of an item to retrieve it. This makes lists suitable for ordered, sequential data where the position of each element matters. In contrast, a dictionary retrieves values using unique keys instead of positions. Each value is associated with a specific key, allowing fast lookups without knowing where the item is stored. Internally, dictionaries use hashing, so retrieving a value by its key is very efficient, typically taking constant time. Because of this key-based access, dictionaries are ideal when you need quick retrieval using meaningful identifiers like names, IDs, or labels, while lists work better when the data is arranged in a natural sequence.
