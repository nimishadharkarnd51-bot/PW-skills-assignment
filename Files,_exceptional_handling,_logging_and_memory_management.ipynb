#15 How does Python's garbage collection system work
Answer- Pythonâ€™s garbage collection system is responsible for automatically managing memory by freeing objects that are no longer needed. This prevents memory leaks and improves performance.

Python mainly uses two mechanisms:

âœ… 1. Reference Counting (Primary Mechanism)

Every object in Python has a reference countâ€”a number that tells how many variables refer to that object.

How it works:

When a new reference is created â†’ count increases

When a reference is deleted â†’ count decreases

When the count becomes zero, Python automatically deletes the object

Example:
import sys

a = [1, 2, 3]
b = a  # reference count increases
del a  # reference count decreases
del b  # reference count becomes 0 â†’ object removed

âœ… 2. Garbage Collector for Circular References

Reference counting fails when two objects reference each other (a circular reference).

Example:

class A:
    def __init__(self):
        self.ref = None

a1 = A()
a2 = A()
a1.ref = a2
a2.ref = a1   # circular reference


Even if a1 and a2 are deleted, their reference count may not become zero.

ðŸ” Pythonâ€™s automatic garbage collector (GC) handles this.

Garbage Collector:

Periodically checks for unreachable objects

Detects cyclic references

Frees them even if reference count > 0

It uses generational garbage collection, dividing objects into three generations:

Generation	Description
  0	        Youngest objects; collected most often
  1	        Surviving objects from generation 0
  2	        Oldest objects; rarely collected

Objects that survive multiple collections are promoted to older generations.

```python
#3 How would you handle a case where the file doesn't exist while trying to open it for reading
#We can handle the case where a file does not exist by using a tryâ€“except block in Python.
#Specifically, we catch the FileNotFoundError exception.
filename = "myfile.txt"

try:
    with open(filename, "r") as file:
        for line in file:
            print(line.strip())
except FileNotFoundError:
    print("Error: The file does not exist.")
```

```python
#4 Write a Python script that reads from one file and writes its content to another file
# Python script to copy contents from one file to another

source_file = "source.txt"
destination_file = "destination.txt"

try:
    # Open source file for reading
    with open(source_file, "r") as src:
        content = src.read()

    # Open destination file for writing
    with open(destination_file, "w") as dest:
        dest.write(content)

    print("File copied successfully!")

except FileNotFoundError:
    print("Error: Source file does not exist.")
```

```python
#5 How would you catch and handle division by zero error in Python
#We can catch and handle a division by zero error in Python using a try-except block
try:
    a = 10
    b = 0
    result = a / b
    print("Result:", result)

except ZeroDivisionError:
    print("Error: You cannot divide a number by zero.")
```

```python
#6 Write a Python program that logs an error message to a log file when a division by zero exception occurs
import logging

# Configure logging
logging.basicConfig(filename="error_log.log",
                    level=logging.ERROR,
                    format="%(asctime)s - %(levelname)s - %(message)s")

try:
    a = 10
    b = 0
    result = a / b

except ZeroDivisionError:
    logging.error("Attempted to divide by zero.")

print("Program completed.")
```

```python
#9 How can you read a file line by line and store its content in a list in Python
# Reading a file line by line and storing in a list

lines = []  # List to store lines

try:
    with open("example.txt", "r") as file:
        for line in file:
            lines.append(line.strip())  # strip() removes \n from each line

    # Printing the stored lines
    print("File contents:")
    for line in lines:
        print(line)

except FileNotFoundError:
    print("Error: The file 'example.txt' does not exist.")
```

```python
#15 Write a Python program that prints the content of a file and handles the case when the file is empty
import os

filename = "example.txt"

if os.path.exists(filename):
    with open(filename, "r") as file:
        content = file.read()
        if content.strip() == "":
            print("The file is empty.")
        else:
            print("File Contents:")
            print(content)
else:
    print("File does not exist.")

```

```python
#16 Demonstrate how to use memory profiling to check the memory usage of a small program
import tracemalloc

# Start memory tracking
tracemalloc.start()

# Small program whose memory we want to check
numbers = [i for i in range(10000)]
total = sum(numbers)

# Get memory usage
current, peak = tracemalloc.get_traced_memory()

print(f"Current memory usage: {current / 1024:.2f} KB")
print(f"Peak memory usage: {peak / 1024:.2f} KB")

# Stop memory tracking
tracemalloc.stop()
```

```python
#18 How would you implement a basic logging setup that logs to a file with rotation after 1MB
import logging
from logging.handlers import RotatingFileHandler

handler = RotatingFileHandler("mylog.log", maxBytes=1024*1024, backupCount=1)
logging.basicConfig(level=logging.INFO, handlers=[handler], format="%(message)s")

logging.info("Hello, this is a log message.")
print("Logging done.")




```

```python
#21 Write a Python program that reads a file and prints the number of occurrences of a specific word
# Program to count occurrences of a word in a file

filename = "sample.txt"
word_to_count = "python"

# Create the file with some content so output always comes
with open(filename, "w") as file:
    file.write("Python is great. I love python programming.\n")

# Now read the file and count the word
with open(filename, "r") as file:
    content = file.read().lower()
    words = content.split()
    count = words.count(word_to_count.lower())

print(f"The word '{word_to_count}' occurs {count} times in the file.")
```

```python
# 22 How can you check if a file is empty before attempting to read its contents
import os

filename = "example.txt"

# Create the file with some content so output always comes
with open(filename, "w") as f:
    f.write("Hello, this file has some text.")

# Check if the file is empty
if os.path.getsize(filename) == 0:
    print("The file is empty.")
else:
    with open(filename, "r") as f:
        print("File is not empty. Contents:")
        print(f.read())
```
