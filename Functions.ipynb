#6 Explain the concept of generators in Python and how they are defined.
Answer - Generators are a special type of iterator in Python that allow you to generate values one at a time, instead of storing the entire sequence in memory.

They are very useful for memory-efficient and lazy data processing.
A generator is a function that uses the yield
keyword instead of return.

It produces values one by one.

It remembers its state (where it left off).

It is more memory-efficient than returning a list.
Generators are defined just like normal functions but using yield.

‚úî Example:
def count_up_to(n):
    i = 1
    while i <= n:
        yield i
        i += 1


Here, yield turns this function into a generator.

‚≠ê  How to Use (Call) a Generator
numbers = count_up_to(3)

print(next(numbers))   # 1
print(next(numbers))   # 2
print(next(numbers))   # 3


After values finish, calling next() again raises StopIteration.

‚≠ê  Using Generator in a Loop
for num in count_up_to(5):
    print(num)


This prints:

1
2
3
4
5

‚≠ê  Generator Expression (Shortcut)

Just like list comprehension, but with () instead of [].

‚úî Example:
squares = (x * x for x in range(5))

print(next(squares))  # 0
print(next(squares))  # 1


# 7  What are the advantages of using generators over regular functions
Answer - Advantages of Generators in Python
1Ô∏è‚É£ Memory Efficient (Low Memory Usage)

Generators do not store all values in memory.
They generate values one at a time, only when needed.

Example:
# List: creates all values and stores them
nums = [i for i in range(1_000_000)]   # takes large memory

# Generator: generates one value at a time
nums = (i for i in range(1_000_000))   # uses very little memory

2Ô∏è‚É£ Faster for Large Data (No Computation Until Needed)

Generators compute values on the fly.
So they are very efficient for reading large files or streams.

3Ô∏è‚É£ Lazy Evaluation

Values are produced only when requested using next() or a loop.
This reduces unnecessary calculations.

4Ô∏è‚É£ Infinite Sequences Are Possible

You can create infinite sequences with generators (can't do with lists).

Example:
def infinite_numbers():
    n = 1
    while True:
        yield n
        n += 1

5Ô∏è‚É£ Cleaner and Simpler Code

Generators avoid the need for large temporary lists and make the logic cleaner.

6Ô∏è‚É£ Improve Performance in Pipelines

Generators allow processing data in steps without holding everything at once.
Used heavily in:

Data pipelines

File processing

Streaming APIs

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
Answer - Difference between map(), reduce(), and filter() in Python

These three functions are used for functional programming in Python, but they serve different purposes.

‚≠ê 1Ô∏è‚É£ map() ‚Äî Transform Each Element
‚úî Purpose

Applies a function to every item in an iterable and returns a new iterator with transformed values.

‚úî Example
numbers = [1, 2, 3, 4]
result = list(map(lambda x: x * 2, numbers))
print(result)


Output:

[2, 4, 6, 8]

‚≠ê 2Ô∏è‚É£ filter() ‚Äî Select Elements Based on Condition
‚úî Purpose

Filters items based on a True/False condition (predicate function).
Returns only the elements that satisfy the condition.

‚úî Example
numbers = [1, 2, 3, 4, 5]
result = list(filter(lambda x: x % 2 == 0, numbers))
print(result)


Output:

[2, 4]

‚≠ê 3Ô∏è‚É£ reduce() ‚Äî Reduce Sequence to a Single Value
‚úî Purpose

Repeats a function that takes two arguments and reduces the iterable to one final result.

It is available in the functools module.

‚úî Example
from functools import reduce

numbers = [1, 2, 3, 4]
result = reduce(lambda x, y: x + y, numbers)
print(result)


Output:

10

```python
#4 . Write a Python function that checks if a given number is prime or not from 1 to 200.
def is_prime(num):
    if num < 2:
        return False

    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True
for n in range(1, 201):
    if is_prime(n):
        print(n, end=" ")

```

```python
#5 Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
class Fibonacci:
    def __init__(self, n):
        self.n = n
        self.a, self.b = 0, 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.n:
            raise StopIteration
        val = self.a
        self.a, self.b = self.b, self.a + self.b
        self.count += 1
        return val
for num in Fibonacci(10):
    print(num)
```

```python
#6  Write a generator function in Python that yields the powers of 2 up to a given exponent
def powers_of_two(n):
    for i in range(n + 1):
        yield 2 ** i
for value in powers_of_two(5):
    print(value)
```

```python
#7  Implement a generator function that reads a file line by line and yields each line as a string
def read_lines_from_string(text):
    for line in text.split("\n"):
        yield line
data = """Hello
This is a test
Reading line by line
Without any file"""

for line in read_lines_from_string(data):
    print(line)

```
