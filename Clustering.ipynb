#2 How does K-Means clustering algorithm work?
Answer : K-Means is an unsupervised clustering algorithm that groups data into K distinct clusters based on similarity, where each data point belongs to the cluster with the nearest mean (centroid).
Step-by-step working of K-Means
Step 1: Choose the number of clusters (K)

You decide how many clusters you want.

Example:
K=3

Step 2: Initialize centroids

Randomly select K data points as initial cluster centroids.

Poor initialization can lead to suboptimal results.

Step 3: Assign points to the nearest centroid

For each data point, calculate the distance to each centroid (usually Euclidean distance).

Assign the point to the cluster with the minimum distance.
Distance= $\sum (x_i - c_i)^2$
Step 4: Update centroids

Recalculate each centroid as the mean of all points assigned to that cluster.
Step 5: Repeat steps 3 and 4

Continue reassigning points and updating centroids until:

Centroids do not change, or

Point assignments remain the same, or

Maximum iterations are reached

#4 What is the main difference between K-Means and Hierarchical Clustering
Answer : The main difference between K-Means and Hierarchical Clustering lies in how clusters are formed and how the number of clusters is handled.

| Type               | Partition-based                             | Tree-based                                                 |

| Cluster shape      | Works best for **spherical clusters**       | Can handle **irregular shapes** better                     |


#5 What are the advantages of DBSCAN over K-Means
Answer : DBSCAN (Density-Based Spatial Clustering of Applications with Noise) has several important advantages over K-Means, especially when dealing with complex, real-world data.
Advantages of DBSCAN over K-Means
1. No need to specify the number of clusters

DBSCAN automatically finds the number of clusters based on data density.

K-Means requires choosing K beforehand.

2. Handles noise and outliers effectively

DBSCAN explicitly labels noise points that do not belong to any cluster.

K-Means assigns every point to a cluster, even outliers.

3. Can find arbitrarily shaped clusters

DBSCAN works well with non-spherical and irregular shapes.

K-Means performs best with spherical, equally sized clusters.

4. More robust to outliers

Outliers do not affect cluster formation significantly in DBSCAN.

In K-Means, outliers can shift centroids, degrading performance.

5. Based on local density, not distance to a centroid

DBSCAN groups points based on dense regions.

K-Means relies on distance to cluster centroids, which can be misleading for complex shapes.

Clusters have irregular shapes

#7 What are the limitations of Hierarchical Clustering
Answer : Hierarchical clustering is a powerful technique, but it has several important limitations, especially when applied to large or complex datasets.
1. Computationally expensive

Time complexity is typically O(n¬≤) or worse.

Not suitable for large datasets.

2. High memory usage

Early mistakes cannot be corrected.

Outliers can distort the clustering structure.

Single linkage is especially prone to the chaining effect.

5. Choice of linkage affects results

Different linkage methods (single, complete, average, Ward) can produce very different dendrograms.

No universal best linkage method.

6. Difficult to interpret for large datasets

Dendrograms become crowded and unreadable as data size grows.

7. Requires a distance metric

Performance depends heavily on the chosen distance measure.

Poor distance choice leads to poor clustering.

#16 Why might K-Means clustering perform poorly on data with varying cluster sizes or densities
Answer : K-Means clustering can perform poorly on data with varying cluster sizes or densities because of how the algorithm works‚Äîit assumes clusters are roughly spherical and equally sized and assigns points based purely on distance to centroids.
Reasons for poor performance
1. Assumes equal cluster sizes

K-Means minimizes inertia, which favors similar-sized clusters.

If one cluster is much larger than another, small clusters may be merged incorrectly or split.

2. Sensitive to varying densities

K-Means uses distance to centroid, not density.

Dense clusters may dominate sparse clusters, causing misassignment of points.

3. Spherical cluster assumption

K-Means assumes clusters are round (isotropic) in all directions.

Non-spherical or elongated clusters are poorly captured.

4. Outliers and noise

Outliers can shift centroids, especially in sparse clusters, affecting cluster assignment.

#17 What are the core parameters in DBSCAN, and how do they influence clustering
Answer : In DBSCAN (Density-Based Spatial Clustering of Applications with Noise), the clustering results are primarily controlled by two core parameters: eps and min_samples. These parameters define what constitutes a dense region, which in turn determines cluster formation and noise detection.
1. eps (Œµ) ‚Äì Neighborhood radius

Defines the maximum distance between two points for them to be considered neighbors.

Influence on clustering:

Small eps ‚Üí many points considered as noise; clusters may be fragmented.

Large eps ‚Üí points far apart may be included; clusters may merge incorrectly.

Essentially controls the scale of density used to detect clusters.

2. min_samples ‚Äì Minimum points for a dense region

Specifies the minimum number of points required to form a core point.

Influence on clustering:

Small min_samples ‚Üí more points become core points; clusters may be too sensitive to noise.

Large min_samples ‚Üí only very dense regions form clusters; sparse regions may be labeled as noise.

3. Derived point types

Core points: Have at least min_samples neighbors within eps.

Border points: Fewer than min_samples neighbors but within eps of a core point.

Noise points: Neither core nor border; lie in sparse regions.

The combination of eps and min_samples controls which points are core, border, or noise, directly influencing cluster size, number, and shape.

Parameter selection tips

Use a k-distance graph (plot distance to k-th nearest neighbor) to find a suitable eps.

Set min_samples based on expected cluster density or at least dimensionality + 1.

```python
#29 Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display as a bar chart
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Generate synthetic data
X, _ = make_blobs(n_samples=500, centers=4, cluster_std=1.0, random_state=42)

# Evaluate silhouette scores for k = 2 to 5
k_values = range(2, 6)
silhouette_scores = []

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X)
    score = silhouette_score(X, labels)
    silhouette_scores.append(score)

# Plot bar chart
plt.figure()
plt.bar(k_values, silhouette_scores)
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Scores for k = 2 to 5")
plt.show()
```

```python
#31 Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with decision boundaries
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic data with overlapping clusters
X, _ = make_blobs(
    n_samples=400,
    centers=3,
    cluster_std=2.0,  # higher std ‚Üí overlapping clusters
    random_state=42
)

# Apply K-Means
kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(X)
centers = kmeans.cluster_centers_

# Create mesh grid for decision boundaries
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(
    np.linspace(x_min, x_max, 300),
    np.linspace(y_min, y_max, 300)
)

# Predict cluster for each point in the grid
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot clusters and decision boundaries
plt.figure()
plt.contourf(xx, yy, Z, alpha=0.3)
plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.scatter(centers[:, 0], centers[:, 1], marker='X', s=200)
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("K-Means with Decision Boundaries (Overlapping Clusters)")
plt.show()
```

```python
#34 Load the Breast Cancer dataset and compare inertia values for K = 2 to 6 using K-Means. Show results in a line plot
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load the Breast Cancer dataset
data = load_breast_cancer()
X = data.data

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Calculate inertia for different K values
k_values = range(2, 7)
inertia = []

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

# Plot inertia values
plt.figure()
plt.plot(k_values, inertia, marker='o')
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Inertia")
plt.title("Elbow Method: Inertia vs K (Breast Cancer Dataset)")
plt.show()
```

```python
#44 Load the Iris dataset and plot the Silhouette Coefficient for each sample after KMeans clustering
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import numpy as np

# Load the Iris dataset
iris = load_iris()
X = iris.data

# Scale features (recommended for K-Means)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(X_scaled)

# Compute Silhouette Coefficient for each sample
silhouette_vals = silhouette_samples(X_scaled, labels)

# Plot Silhouette Coefficient for each sample
plt.figure(figsize=(8, 5))
y_lower = 0

for i in np.unique(labels):
    ith_cluster_silhouette = silhouette_vals[labels == i]
    ith_cluster_silhouette.sort()
    size_cluster = ith_cluster_silhouette.shape[0]
    y_upper = y_lower + size_cluster

    plt.barh(
        range(y_lower, y_upper),
        ith_cluster_silhouette
    )
    y_lower = y_upper

plt.xlabel("Silhouette Coefficient")
plt.ylabel("Samples")
plt.title("Silhouette Coefficient for Each Sample (Iris Dataset)")
plt.show()
```

```python
#47 Generate noisy blobs using make_blobs and use DBSCAN to identify both clusters and noise points. Print the count
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import numpy as np

# Generate noisy blob data
X, _ = make_blobs(
    n_samples=500,
    centers=3,
    cluster_std=1.2,
    random_state=42
)

# Add some random noise points
rng = np.random.RandomState(42)
noise = rng.uniform(low=-10, high=10, size=(50, 2))
X_noisy = np.vstack([X, noise])

# Scale features (important for DBSCAN)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_noisy)

# Apply DBSCAN
dbscan = DBSCAN(eps=0.4, min_samples=5)
labels = dbscan.fit_predict(X_scaled)

# Count clusters (excluding noise)
num_clusters = len(set(labels)) - (1 if -1 in labels else 0)

# Count noise points
num_noise = np.sum(labels == -1)

print("Number of clusters identified:", num_clusters)
print("Number of noise points identified:", num_noise)
```
