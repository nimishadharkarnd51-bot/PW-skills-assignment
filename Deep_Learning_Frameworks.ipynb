#3 What is the primary function of the tf.function in TensorFlow 2.0
Answer : The primary function of tf.function in TensorFlow 2.0 is to convert regular Python functions into optimized TensorFlow computation graphs. This allows code written in an intuitive, eager-execution style to run faster and more efficiently, similar to TensorFlow 1.x graphs, while still keeping the simplicity of TensorFlow 2.0.

In practice, tf.function traces the Python function and builds a static graph that can be optimized, parallelized, and executed on different hardware (CPU, GPU, TPU). This improves performance, scalability, and deployability without requiring users to manually define graphs or sessions.

#6 What is the importance of Tensor Space in TensorFlow

Overall, tensor space provides the mathematical and computational foundation of TensorFlow, making it possible to build scalable, high-performance, and flexible deep learning models.

#10 What is the difference between TensorFlow and PyTorch
Answer : TensorFlow and PyTorch are two of the most popular deep learning frameworks, but they differ mainly in computation style, usability, and ecosystem focus.

TensorFlow (especially TensorFlow 2.x) uses a high-level Keras API and supports both eager execution and graph execution through tf.function. It is highly optimized for production deployment, offering strong support for mobile (TensorFlow Lite), web (TensorFlow.js), and large-scale distributed systems, making it a preferred choice in industry.

PyTorch, on the other hand, is known for its dynamic computation graph and Pythonic, intuitive coding style, which makes model development and debugging easier. Its “define-by-run” approach allows the graph to be built dynamically during execution, making it extremely popular in research and experimentation.

In terms of ecosystem, TensorFlow provides comprehensive end-to-end tools such as TensorBoard, TFX, and serving pipelines, while PyTorch relies on a more flexible ecosystem with tools like TorchServe and ONNX. Overall, TensorFlow is often favored for scalable production systems, whereas PyTorch is preferred for research, prototyping, and rapid experimentation.

#12 What is the basic structure of a PyTorch neural network
Answer : The basic structure of a PyTorch neural network is built by defining a model class, specifying layers in the constructor, and defining the forward computation explicitly. PyTorch follows a define-by-run approach, meaning the computation graph is created dynamically during execution.

Basic Structure of a PyTorch Neural Network
1. Import Required Modules
import torch
import torch.nn as nn
import torch.nn.functional as F

2. Define the Model Class

A neural network is defined by inheriting from nn.Module.

class NeuralNet(nn.Module):
    def __init__(self):
        super(NeuralNet, self).__init__()
        self.fc1 = nn.Linear(784, 128)
        self.fc2 = nn.Linear(128, 64)
        self.fc3 = nn.Linear(64, 10)

    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.fc3(x)
        return x

3. Create the Model Object
model = NeuralNet()

4. Define Loss Function and Optimizer
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

5. Training Loop (Simplified)
for data, target in train_loader:
    optimizer.zero_grad()
    output = model(data)
    loss = criterion(output, target)
    loss.backward()
    optimizer.step()

Key Components Explained

nn.Module: Base class for all neural networks

Layers (nn.Linear): Define network structure

forward() method: Defines data flow

Loss function: Measures error

Optimizer: Updates weights

#14 What is the difference between torch.Tensor and torch.cuda.Tensor in PyTorch
ANswer : The difference between torch.Tensor and torch.cuda.Tensor in PyTorch lies in where the tensor is stored and computed.

A torch.Tensor is a general tensor object that, by default, resides on the CPU. It is used for standard numerical computations and model training when GPU acceleration is not required. All operations on a CPU tensor are executed using the system’s CPU.

A torch.cuda.Tensor, on the other hand, is a tensor that resides on the GPU (CUDA device). It enables parallel computation, making operations significantly faster for large-scale matrix and deep learning workloads. To create a CUDA tensor, a CPU tensor is typically moved to the GPU using .cuda() or .to('cuda').

In practice, both are the same tensor type logically, but the key difference is the device on which the tensor lives. PyTorch automatically dispatches operations to CPU or GPU based on the tensor’s device, and tensors must be on the same device to interact in computations.

Variants of ReLU with smoother gradients

#17 What is the difference between torch.nn.Module and torch.nn.Sequential in PyTorch
Answer : The difference between torch.nn.Module and torch.nn.Sequential in PyTorch lies in their flexibility and use cases.

torch.nn.Module is the base class for all neural network models in PyTorch. It allows you to define custom architectures by explicitly specifying layers and writing a custom forward() method. This makes it highly flexible and suitable for complex models involving branching, skip connections, multiple inputs/outputs, or conditional logic.

torch.nn.Sequential, on the other hand, is a container module that automatically applies layers in a linear, sequential order. It is ideal for simple feedforward networks where data flows straight from one layer to the next without any branching or custom operations. With nn.Sequential, you do not need to explicitly define a forward() method.

In summary, nn.Module offers full control and flexibility, while nn.Sequential provides simplicity and convenience for straightforward, stack-like model architectures.

```python
#3 How can you create a simple neural network in TensorFlow 2.0 with one hidden laye
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Create the model
model = Sequential([
    Dense(16, activation='relu', input_shape=(10,)),  # Hidden layer
    Dense(1, activation='sigmoid')                     # Output layer
])

# Compile the model
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

# Display model summary
model.summary()
```

```python
#4 How can you visualize the training progress using TensorFlow and Matplotlib
import tensorflow as tf
import matplotlib.pyplot as plt

# Example model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(16, activation='relu', input_shape=(10,)),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

# Dummy data
import numpy as np
X = np.random.rand(200, 10)
y = np.random.randint(0, 2, 200)

# Train the model
history = model.fit(X, y, epochs=10, validation_split=0.2, verbose=0)

# Plot training & validation loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend(['Training Loss', 'Validation Loss'])
plt.show()

# Plot training & validation accuracy
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend(['Training Accuracy', 'Validation Accuracy'])
plt.show()
```
